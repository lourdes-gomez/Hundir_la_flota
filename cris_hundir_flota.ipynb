```python
tablero = pd.DataFrame(np.full((10,10), ' '))
tablero
```

```python
barcos = {
    'barco1_1' : {
        'cantidad' : 1,
        'vidas' : 1},
    'barco1_2' : {
        'cantidad' : 1,
        'vidas' : 1},
    'barco1_3' : {
        'cantidad' : 1,
        'vidas' : 1},
    'barco1_4' : {
        'cantidad' : 1,
        'vidas' : 1},
    'barco2_1' : {
        'cantidad' : 1,
        'vidas' : 2},
    'barco2_2' : {
        'cantidad' : 1,
        'vidas' : 2},
    'barco2_3' : {
        'cantidad' : 1,
        'vidas' : 2},
    'barco3_1' : {
        'cantidad' : 1,
        'vidas' : 3},
    'barco3_2' : {
        'cantidad' : 1,
        'vidas' : 3},
    'barco4_1' : {
        'cantidad' : 1,
        'vidas' : 4}}
```

```python
def espacio_barco(c1,c2,orientacion,eslora):
    if orientacion == 'h':
        espacio_barco = tablero.loc[c1-1:c1+1, c2-1:c2-1+eslora+2] 
    if orientacion == 'v':
        espacio_barco = tablero.loc[c1-1:c1+eslora+1,c2-1:c2+1]
    return espacio_barco
```

```python
def colocar_barco(eslora):
    lista_barcos = []
    sigo = True
    while sigo:
        sigo = False
        c1 = np.random.randint(0,10)
        c2 = np.random.randint(0,10)
        orientacion = np.random.choice(['h', 'v'])

        if orientacion == 'h' and ((c2+eslora)<10):
            
            if np.any(espacio_barco(c1,c2,orientacion,eslora) == 'O'):
                sigo = True
            else:
                tablero.loc[c1, c2:c2+eslora] = 'O'
                barco = [(c1+eslora,c2),(c1,c2)]
                lista_barcos.append(barco)

        elif orientacion == 'v' and ((c1+eslora)<10):
            
            if np.any(espacio_barco(c1,c2,orientacion,eslora) == 'O'):
                sigo = True
            else: 
                tablero.loc[c1:c1+eslora, c2] = 'O'
                barco = [(c1,c2),(c1+eslora,c2)]
                lista_barcos.append(barco)
 
        else:
            sigo = True
    return lista_barcos
```

```python
def tocado_o_hundido(c1,c2, vidas):
    if c1 < 6 and (tablero[c1+1,c2] == 'O' or tablero[c1+2,c2] == 'O' or tablero[c1+3,c2]) == 'O': #barco hacia abajo
        print('Tocado')
        vidas = vidas -1
    elif c2 < 6 and (tablero[c1,c2+1] == 'O' or tablero[c1,c2+2] == 'O' or tablero[c1,c2+3]) == 'O': #barco hacia la derecha
        print('Tocado')
        vidas = vidas - 1
    elif  c1 > 3 and (tablero[c1-1,c2] == 'O' or tablero[c1-2,c2] == 'O' or tablero[c1-3,c2]) == 'O': #barco hacia arriba
        print('Tocado')
        vidas = vidas - 1
    elif  c2 > 3 and (tablero[c1,c2-1] == 'O' or tablero[c1,c2-2] == 'O' or tablero[c1,c2-3]) == 'O': #barco hacia la izquierda
        print('Tocado')
        vidas = vidas - 1
    tablero[c1,c2] = 'X'
    return vidas
```

```python
def disparo(tablero, vidas, lista_b):
    #turno = 'yo'
    sigue_disparando = True
    while sigue_disparando == True:
        sigue_disparando = False
        c1 = int(input('Coordenada 1 '))
        c2 = int(input('Coordenada 2 '))
        if tablero.loc[c1,c2] == ' ':
            print('Agua')
            #turno = 'maquina'
            #print(f'Le toca el turno a {turno}')
            return(tablero)
        elif tablero.loc[c1, c2] == 'O':
            for i in range(len(lista_b)):
                if lista_b[i][0] == (c1,c2) and lista_b[i][0] == lista_b[i][1]:  # c1,c2 coincide con la primera coordenada de un barco
                    #if lista_b[i][0] == lista_b[i][1]: # Barco de eslora 1(sera hundido)
                    print('Hundido')
                    tablero.loc[c1,c2] = 'X'
                    #lista_b.remove([c1,c2],[c1,c2])
                    vidas = vidas -1
                    print(tablero)
                else:        
                    vidas = tocado_o_hundido(c1,c2,vidas)
        sigue_disparando = True
        print (vidas)
        print (tablero)
```
